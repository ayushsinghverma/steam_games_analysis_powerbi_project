```python
def e1(i):
    for i in df1['Estimated owners']:
        if i=='0 - 20000':
            return 0
        elif i=='20000 - 50000':
            return 1
        elif i=='0 - 0':
            return 2
        elif i=='50000 - 100000':
            return 3
        elif i=='100000 - 200000':
            return 4
        elif i=='200000 - 500000':
            return 5
        elif i=='500000 - 1000000':
            return 6
        elif i=='1000000 - 2000000':
            return 7
        elif i=='2000000 - 5000000':
            return 8
        elif i=='5000000 - 10000000':
            return 9
        elif i=='10000000 - 20000000':
            return 10
        elif i=='20000000 - 50000000':
            return 11
        elif i=='50000000 - 100000000':
            return 12
        elif i=='100000000 - 200000000':
            return 13
```
